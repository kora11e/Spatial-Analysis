```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import geopandas as geon
```

```python
df = geon.read_file('./geoData/districtData/district.shp')
```

```python
df.head(3)
```

```python
df.plot()
```

```python
df.plot('Shape_Area', cmap='BrBG')
```

```python
df.plot('SUM_area', cmap='hsv', legend=True, vmin=1000, vmax=6000, edgecolor='black', legend_kwds={'label': 'Sum Area'})
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))

df.plot('SUM_area', cmap='ocean', ax=ax, legend=True, vmin=1000, vmax=6000, edgecolor='black', legend_kwds={'label': 'Sum Area'})

ax.set_title('Nepal Area')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_xticks([])
ax.set_yticks([])
```

```python
provinces = df['Province'].unique()
provinces
```

```python
fig,ax = plt.subplots(2, 4, figsize=(15,10))
axis = ax.flatten()

for (ax, province) in zip(axis, provinces):
    df_province = df[df['Province']==province]

    df_province.plot('SUM_area', cmap = 'ocean', ax=ax, edgecolor='black')

    ax.set_title('Nepal Area')
    ax.set_xticks([])
    ax.set_yticks([])
```
